```python
import os

folder = r"C:\Users\dell\Untitled Folder 5"
os.chdir(folder)

# ── index.html ───────────────────────────
index_html = '''<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Support CRM</title>
  <script src="https://cdn.tailwindcss.com"></script>
</head>
<body class="bg-gray-100 min-h-screen">
  <nav class="bg-blue-700 text-white px-6 py-4 flex justify-between items-center">
    <h1 class="text-xl font-bold">🎫 Support CRM</h1>
    <a href="create.html" class="bg-white text-blue-700 px-4 py-2 rounded font-bold">+ New Ticket</a>
  </nav>
  <div class="max-w-6xl mx-auto p-6">
    <div class="grid grid-cols-4 gap-4 mb-6">
      <div class="bg-white rounded shadow p-4 text-center border-t-4 border-blue-500">
        <p class="text-gray-500 text-sm">Total</p>
        <p class="text-3xl font-bold text-blue-600" id="total">0</p>
      </div>
      <div class="bg-white rounded shadow p-4 text-center border-t-4 border-red-500">
        <p class="text-gray-500 text-sm">Open</p>
        <p class="text-3xl font-bold text-red-500" id="open_count">0</p>
      </div>
      <div class="bg-white rounded shadow p-4 text-center border-t-4 border-yellow-500">
        <p class="text-gray-500 text-sm">In Progress</p>
        <p class="text-3xl font-bold text-yellow-500" id="prog_count">0</p>
      </div>
      <div class="bg-white rounded shadow p-4 text-center border-t-4 border-green-500">
        <p class="text-gray-500 text-sm">Closed</p>
        <p class="text-3xl font-bold text-green-500" id="closed_count">0</p>
      </div>
    </div>
    <div class="flex gap-3 mb-4">
      <input id="search" type="text" placeholder="Search name, email, ID..."
        oninput="load()"
        class="flex-1 border rounded px-4 py-2 focus:outline-none focus:ring-2 focus:ring-blue-400"/>
      <select id="sf" onchange="load()"
        class="border rounded px-4 py-2">
        <option value="">All Status</option>
        <option value="Open">Open</option>
        <option value="In Progress">In Progress</option>
        <option value="Closed">Closed</option>
      </select>
    </div>
    <div class="bg-white rounded shadow overflow-hidden">
      <table class="w-full text-sm">
        <thead class="bg-gray-50 text-gray-500 uppercase text-xs">
          <tr>
            <th class="px-4 py-3 text-left">ID</th>
            <th class="px-4 py-3 text-left">Customer</th>
            <th class="px-4 py-3 text-left">Subject</th>
            <th class="px-4 py-3 text-left">Status</th>
            <th class="px-4 py-3 text-left">Date</th>
            <th class="px-4 py-3 text-left">Action</th>
          </tr>
        </thead>
        <tbody id="tbl"></tbody>
      </table>
    </div>
  </div>
  <script>
    const API = "http://localhost:8000";
    function badge(s){
      if(s==="Open") return \'<span class="bg-red-100 text-red-600 px-2 py-1 rounded-full text-xs font-bold">Open</span>\';
      if(s==="In Progress") return \'<span class="bg-yellow-100 text-yellow-600 px-2 py-1 rounded-full text-xs font-bold">In Progress</span>\';
      return \'<span class="bg-green-100 text-green-600 px-2 py-1 rounded-full text-xs font-bold">Closed</span>\';
    }
    async function load(){
      const s=document.getElementById("search").value;
      const f=document.getElementById("sf").value;
      let url=`${API}/api/tickets?`;
      if(s) url+=`search=${s}&`;
      if(f) url+=`status=${f}`;
      try{
        const r=await fetch(url);
        const d=await r.json();
        document.getElementById("total").textContent=d.length;
        document.getElementById("open_count").textContent=d.filter(t=>t.status==="Open").length;
        document.getElementById("prog_count").textContent=d.filter(t=>t.status==="In Progress").length;
        document.getElementById("closed_count").textContent=d.filter(t=>t.status==="Closed").length;
        document.getElementById("tbl").innerHTML=d.length===0
          ? \'<tr><td colspan="6" class="text-center py-8 text-gray-400">No tickets found.</td></tr>\'
          : d.map(t=>`<tr class="border-t hover:bg-gray-50">
              <td class="px-4 py-3 font-bold text-blue-600">${t.ticket_id}</td>
              <td class="px-4 py-3"><div class="font-medium">${t.customer_name}</div>
              <div class="text-gray-400 text-xs">${t.customer_email}</div></td>
              <td class="px-4 py-3">${t.subject}</td>
              <td class="px-4 py-3">${badge(t.status)}</td>
              <td class="px-4 py-3 text-gray-400 text-xs">${new Date(t.created_at).toLocaleDateString()}</td>
              <td class="px-4 py-3"><a href="detail.html?id=${t.ticket_id}"
                class="bg-blue-600 text-white px-3 py-1 rounded text-xs font-bold">View</a></td>
            </tr>`).join("");
      }catch(e){
        document.getElementById("tbl").innerHTML=\'<tr><td colspan="6" class="text-center py-8 text-red-400">Server not running!</td></tr>\';
      }
    }
    load();
  </script>
</body>
</html>'''
```

```python
create_html = '''<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Create Ticket</title>
  <script src="https://cdn.tailwindcss.com"></script>
</head>
<body class="bg-gray-100 min-h-screen">
  <nav class="bg-blue-700 text-white px-6 py-4 flex justify-between items-center">
    <h1 class="text-xl font-bold">🎫 Support CRM</h1>
    <a href="index.html" class="bg-white text-blue-700 px-4 py-2 rounded font-bold">← Back</a>
  </nav>
  <div class="max-w-xl mx-auto p-6">
    <h2 class="text-2xl font-bold mb-6">Create New Ticket</h2>
    <div id="err" class="bg-red-50 border border-red-300 text-red-600 rounded px-4 py-3 mb-4 hidden"></div>
    <div id="ok" class="bg-green-50 border border-green-300 text-green-600 rounded px-4 py-3 mb-4 hidden"></div>
    <div class="bg-white rounded shadow p-6 space-y-4">
      <div class="grid grid-cols-2 gap-4">
        <div>
          <label class="block text-sm font-semibold text-gray-600 mb-1">Name *</label>
          <input id="nm" type="text" placeholder="John Doe"
            class="w-full border rounded px-3 py-2 focus:outline-none focus:ring-2 focus:ring-blue-400"/>
        </div>
        <div>
          <label class="block text-sm font-semibold text-gray-600 mb-1">Email *</label>
          <input id="em" type="email" placeholder="john@gmail.com"
            class="w-full border rounded px-3 py-2 focus:outline-none focus:ring-2 focus:ring-blue-400"/>
        </div>
      </div>
      <div>
        <label class="block text-sm font-semibold text-gray-600 mb-1">Subject *</label>
        <input id="sb" type="text" placeholder="Issue title"
          class="w-full border rounded px-3 py-2 focus:outline-none focus:ring-2 focus:ring-blue-400"/>
      </div>
      <div>
        <label class="block text-sm font-semibold text-gray-600 mb-1">Description *</label>
        <textarea id="ds" rows="4" placeholder="Describe the issue..."
          class="w-full border rounded px-3 py-2 focus:outline-none focus:ring-2 focus:ring-blue-400"></textarea>
      </div>
      <button onclick="submit()"
        class="w-full bg-blue-600 text-white py-2 rounded font-bold hover:bg-blue-700">
        🎫 Create Ticket
      </button>
    </div>
  </div>
  <script>
    const API="http://localhost:8000";
    async function submit(){
      const n=document.getElementById("nm").value.trim();
      const e=document.getElementById("em").value.trim();
      const s=document.getElementById("sb").value.trim();
      const d=document.getElementById("ds").value.trim();
      const err=document.getElementById("err");
      const ok=document.getElementById("ok");
      err.classList.add("hidden");
      ok.classList.add("hidden");
      if(!n||!e||!s||!d){
        err.textContent="All fields required!";
        err.classList.remove("hidden");
        return;
      }
      try{
        const r=await fetch(`${API}/api/tickets`,{
          method:"POST",
          headers:{"Content-Type":"application/json"},
          body:JSON.stringify({customer_name:n,customer_email:e,subject:s,description:d})
        });
        const data=await r.json();
        ok.textContent=`Ticket ${data.ticket_id} created! Redirecting...`;
        ok.classList.remove("hidden");
        setTimeout(()=>window.location.href="index.html",2000);
      }catch(err2){
        err.textContent="Server not running!";
        err.classList.remove("hidden");
      }
    }
  </script>
</body>
</html>'''

# ── detail.html ──────────────────────────
detail_html = '''<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Ticket Detail</title>
  <script src="https://cdn.tailwindcss.com"></script>
</head>
<body class="bg-gray-100 min-h-screen">
  <nav class="bg-blue-700 text-white px-6 py-4 flex justify-between items-center">
    <h1 class="text-xl font-bold">🎫 Support CRM</h1>
    <a href="index.html" class="bg-white text-blue-700 px-4 py-2 rounded font-bold">← Back</a>
  </nav>
  <div class="max-w-4xl mx-auto p-6">
    <div id="loading" class="text-center py-12 text-gray-400">Loading...</div>
    <div id="content" class="hidden">
      <div class="flex items-center gap-3 mb-2">
        <span id="tid" class="text-blue-600 font-bold text-lg"></span>
        <span id="sbadge"></span>
      </div>
      <h2 id="tsubj" class="text-2xl font-bold mb-6"></h2>
      <div class="grid grid-cols-3 gap-6">
        <div class="col-span-2 space-y-4">
          <div class="bg-white rounded shadow p-5">
            <h3 class="font-bold text-gray-500 text-xs uppercase mb-3">Customer Info</h3>
            <div class="space-y-2 text-sm">
              <div class="flex justify-between border-b pb-2">
                <span class="text-gray-400">Name</span>
                <span id="cn" class="font-medium"></span>
              </div>
              <div class="flex justify-between border-b pb-2">
                <span class="text-gray-400">Email</span>
                <span id="ce" class="font-medium text-blue-600"></span>
              </div>
              <div class="flex justify-between">
                <span class="text-gray-400">Created</span>
                <span id="cc" class="font-medium"></span>
              </div>
            </div>
          </div>
          <div class="bg-white rounded shadow p-5">
            <h3 class="font-bold text-gray-500 text-xs uppercase mb-3">Description</h3>
            <p id="cdesc" class="text-gray-700 text-sm leading-relaxed"></p>
          </div>
          <div class="bg-white rounded shadow p-5">
            <h3 class="font-bold text-gray-500 text-xs uppercase mb-3">Notes</h3>
            <div id="notes"></div>
          </div>
        </div>
        <div class="space-y-4">
          <div class="bg-white rounded shadow p-5">
            <h3 class="font-bold text-gray-500 text-xs uppercase mb-3">Update</h3>
            <label class="block text-sm font-semibold text-gray-600 mb-1">Status</label>
            <select id="ns" class="w-full border rounded px-3 py-2 mb-3">
              <option>Open</option>
              <option>In Progress</option>
              <option>Closed</option>
            </select>
            <label class="block text-sm font-semibold text-gray-600 mb-1">Note</label>
            <textarea id="nn" rows="3" placeholder="Add note..."
              class="w-full border rounded px-3 py-2 mb-3 text-sm"></textarea>
            <div id="umsg" class="text-green-600 text-sm mb-2 hidden">Updated!</div>
            <button onclick="upd()"
              class="w-full bg-blue-600 text-white py-2 rounded font-bold hover:bg-blue-700">
              Save
            </button>
          </div>
          <div class="bg-white rounded shadow p-5 space-y-2">
            <button onclick="cls()"
              class="w-full border border-green-400 text-green-600 py-2 rounded font-bold hover:bg-green-50 text-sm">
              Close Ticket
            </button>
            <button onclick="del()"
              class="w-full border border-red-300 text-red-500 py-2 rounded font-bold hover:bg-red-50 text-sm">
              Delete Ticket
            </button>
          </div>
        </div>
      </div>
    </div>
  </div>
  <script>
    const API="http://localhost:8000";
    const id=new URLSearchParams(window.location.search).get("id");
    function badge(s){
      if(s==="Open") return \'<span class="bg-red-100 text-red-600 px-2 py-1 rounded-full text-xs font-bold">Open</span>\';
      if(s==="In Progress") return \'<span class="bg-yellow-100 text-yellow-600 px-2 py-1 rounded-full text-xs font-bold">In Progress</span>\';
      return \'<span class="bg-green-100 text-green-600 px-2 py-1 rounded-full text-xs font-bold">Closed</span>\';
    }
    async function load(){
      try{
        const r=await fetch(`${API}/api/tickets/${id}`);
        const t=await r.json();
        document.getElementById("loading").classList.add("hidden");
        document.getElementById("content").classList.remove("hidden");
        document.getElementById("tid").textContent=t.ticket_id;
        document.getElementById("sbadge").innerHTML=badge(t.status);
        document.getElementById("tsubj").textContent=t.subject;
        document.getElementById("cn").textContent=t.customer_name;
        document.getElementById("ce").textContent=t.customer_email;
        document.getElementById("cc").textContent=new Date(t.created_at).toLocaleString();
        document.getElementById("cdesc").textContent=t.description;
        document.getElementById("ns").value=t.status;
        const nl=document.getElementById("notes");
        nl.innerHTML=t.notes&&t.notes.length>0
          ? t.notes.map(n=>`<div class="bg-gray-50 border-l-4 border-blue-400 rounded p-3 mb-2">
              <p class="text-sm">${n.note_text}</p>
              <p class="text-xs text-gray-400 mt-1">${new Date(n.created_at).toLocaleString()}</p>
            </div>`).join("")
          : \'<p class="text-gray-400 text-sm">No notes yet.</p>\';
      }catch(e){
        document.getElementById("loading").textContent="Error loading ticket!";
      }
    }
    async function upd(){
      const s=document.getElementById("ns").value;
      const n=document.getElementById("nn").value.trim();
      await fetch(`${API}/api/tickets/${id}`,{
        method:"PUT",
        headers:{"Content-Type":"application/json"},
        body:JSON.stringify({status:s,note:n})
      });
      document.getElementById("umsg").classList.remove("hidden");
      document.getElementById("nn").value="";
      setTimeout(()=>document.getElementById("umsg").classList.add("hidden"),3000);
      load();
    }
    async function cls(){
      if(confirm("Close ticket?")){
        await fetch(`${API}/api/tickets/${id}`,{
          method:"PUT",
          headers:{"Content-Type":"application/json"},
          body:JSON.stringify({status:"Closed",note:"Ticket closed."})
        });
        load();
      }
    }
    async function del(){
      if(confirm("Delete ticket?")){
        await fetch(`${API}/api/tickets/${id}`,{method:"DELETE"});
        window.location.href="index.html";
      }
    }
    load();
  </script>
</body>
</html>'''

# Save all files
with open(os.path.join(folder, "index.html"), "w", encoding="utf-8") as f:
    f.write(index_html)

with open(os.path.join(folder, "create.html"), "w", encoding="utf-8") as f:
    f.write(create_html)

with open(os.path.join(folder, "detail.html"), "w", encoding="utf-8") as f:
    f.write(detail_html)

print("✅ Teeno files ban gayi!")
print(f"📁 Folder: {folder}")
print("Files:", os.listdir(folder))
```
